## Create Sample Data

Let's create a messy CSV with common data quality issues.

```
import csv

rows = [
    ['name', 'phone', 'email', 'date', 'state'],
    ['  John Smith  ', '(555) 123-4567', 'JOHN@EXAMPLE.COM', '03/15/2024', 'California'],
    ['jane doe', '555.987.6543', 'jane@example.com', '2024-03-16', 'CA'],
    ['BOB WILSON', '5551234567', 'Bob@Example.Com', '15-Mar-2024', 'calif.'],
    ['  alice brown', '(555)111-2222', 'ALICE@EXAMPLE.COM', '03/17/24', 'Cal'],
    ['Charlie Davis', '555 333 4444', 'charlie@example.com', 'March 18, 2024', 'california'],
]

with open('messy_data.csv', 'w', newline='') as f:
    csv.writer(f).writerows(rows)

print('Created messy_data.csv with formatting inconsistencies')
```

## Zero-Config Transform

GoldenFlow auto-detects issues and fixes them — no config needed.

```
!goldenflow transform messy_data.csv --output-dir output/
```

```
import polars as pl

print('=== BEFORE ===')
print(pl.read_csv('messy_data.csv'))
print()
print('=== AFTER ===')
print(pl.read_csv('output/messy_data_transformed.csv'))
```
